```python
# View SQLite basket data
# ------------------------

import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect("baskets.db")

# Load and display baskets
baskets_df = pd.read_sql_query("SELECT * FROM baskets", conn)
print("üß∫ Baskets Table:")
print(baskets_df)

# Load and display basket items
items_df = pd.read_sql_query("SELECT * FROM basket_items", conn)
print("\nüì¶ Basket Items Table:")
print(items_df)

# Optionally merge for full view
merged_df = pd.read_sql_query("""
    SELECT b.id as basket_id, b.timestamp, i.item
    FROM baskets b
    JOIN basket_items i ON b.id = i.basket_id
    ORDER BY b.id
""", conn)
print("\nüîç Full Basket Data:")
print(merged_df)

conn.close()
```
